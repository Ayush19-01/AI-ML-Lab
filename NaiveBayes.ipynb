```python
#Ayush Sharma 209303312
# 8.1 Program to demonstrate implementation of na√Øve Bayes Theorem in python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
url=r'https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/titanic.csv'
titanicdf=pd.read_csv(url)
titanicdf.head()
```

```python
df=titanicdf
df.head()
```

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
df['Sex'] = pd.get_dummies(df['Sex'], drop_first=True)
df.head()
```

```python
df.columns[df.isna().any()]
df.Age[:10]
df.Age =df.Age.fillna(df.Age.mean())
df.head()
```

```python
X = df.drop(['Survived'], axis=1)
y = df['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = GaussianNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
